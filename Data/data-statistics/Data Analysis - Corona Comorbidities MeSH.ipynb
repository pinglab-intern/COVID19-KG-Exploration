### Data Stat

This notebook sumarizes the preliminary data found in Corona virus, Covid-19 document count, time published, cohort studied etc.

```python
import json as json
```

```python
with open("corona_comorbidities_id2name.json",'r')as rf:
    id2name = json.load(rf)
```

```python
DATA = []
with open("../mesh_search/mesh_corona_comorbidities_paper_ids.json", 'r') as fcount:
        data = json.load(fcount)
        for key,value in data.items():
            DATA.append({"MeSH":key,"Count":len(value)})
```

```python
DATA
```

```python
import pandas as pd
DF = pd.DataFrame(DATA)
```

```python
DF = DF.sort_values('Count', ascending =False)
```

```python
DF
```

```python
DF['Name'] = DF['MeSH'].apply(lambda x: id2name[x])
```

```python
DF = DF.set_index('Name')
```

```python
DF
```

```python
DF.to_csv('corona_comorbidities_mesh_count.csv')
```
