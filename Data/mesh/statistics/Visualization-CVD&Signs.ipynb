## Data Visualization

We will show the cooccurance data in a visualization using seaborn package

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sns.set()
```

#### Coocurance of Covid 19 Signs and Cardio Vascular Disease Mesh Terms

```python
data_sign_cvd = pd.read_csv("signs_cvd.csv")
```

```python
data_sign_cvd.head(10)
```

```python
C2 = data_sign_cvd['Signs:']
C1 = data_sign_cvd['CVD:']
V = data_sign_cvd['Overlap']
```

```python
U1 = []
for item in C1:
    if item not in U1:
        U1.append(item)
```

```python
Data = []
for e1 in U1:
    data = {"name":e1}
    for ex,e2,v in zip(C1,C2,V):
        if ex==e1:
            data.update({e2:v})
    Data.append(data)
```

```python
DF_sign_cvd = pd.DataFrame(Data)
```

```python
DF_sign_cvd = DF_sign_cvd.set_index("name")
DF_sign_cvd = DF_sign_cvd.fillna(0)
DF_sign_cvd.shape
```

```python
plt.figure(figsize = [12,60])
sns.heatmap(DF_sign_cvd, cmap="YlGnBu", annot = True, fmt='.0f')
```
